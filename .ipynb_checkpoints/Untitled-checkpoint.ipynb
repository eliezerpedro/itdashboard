```python
import json
import os
import shutil
import pandas as pd
from pathlib import Path
from time import sleep
import re
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
```

```python
dirpath = os.getcwd()+'\\chromedriver.exe' 
download_path = os.getcwd()+'\\output'
chrome_options = Options()
chrome_options.add_argument('--start-maximized')
chrome_options.add_argument("--enable-javascript")

appState = {
   "recentDestinations": [
        {
            "id": "Save as PDF",
            "origin": "local",
            "account": ""
        }
    ],
    "selectedDestinationId": "Save as PDF",
    "version": 2
}
prefs = {
    'printing.print_preview_sticky_settings.appState': json.dumps(appState),
    'savefile.default_directory': download_path
        }

chrome_options.add_experimental_option('prefs', prefs)
chrome_options.add_argument('--kiosk-printing')

browser = webdriver.Chrome(executable_path=dirpath, options=chrome_options) #instancia o driver
```

```python
browser.get('https://www.itdashboard.gov/itportfoliodashboard')

agency_dic = {"Agency": [],
       "FY 2022 IT Spending": [],
       "Spending on Major Investments": []
      }
for agency in browser.find_element_by_id('agency-select').text.split('\n'):
    
    browser.find_element_by_xpath(f'*//option[contains(text(), "{agency}")]').click()
    
    while 1:
        try:
            browser.find_element_by_xpath('*//div[@class="spending-overview"]//h3')
            break
        except NoSuchElementException:
            pass
    
    while not browser.find_element_by_xpath('*//div[@class="spending-overview"]//h3').text == agency.upper(): 
        sleep(1)
        
    it_spending = browser.find_element_by_xpath('*//div[@class="it-spending"]').text.split('\n')[1]
    major_spending = browser.find_element_by_xpath('*//div[@class="major-spending"]').text.split('\n')[1]
    
    agency_dic['Agency'].append(agency)
    agency_dic['FY 2022 IT Spending'].append(it_spending)
    agency_dic['Spending on Major Investments'].append(major_spending)

df_agencias = pd.DataFrame(agency_dic)
```

```python

choose_agency = 'Department of Commerce'

#browser.find_elements_by_xpath('*//span[contains(text(), "Advanced Search")]')[1].click()
browser.get('https://www.itdashboard.gov/index.php/search-advanced')
browser.find_element_by_id('edit-keywords').send_keys(choose_agency)
browser.find_element_by_id('edit-submit').click()
while 1:
    try:
        browser.find_element_by_xpath(f'*//option[contains(text(), "40")]').click()
        break
    except:
        sleep(0.1)
while not re.search('40', browser.find_elements_by_xpath('*//div[@class="number-of-results"]')[1].text):
    sleep(1)
pages = int(browser.find_elements_by_xpath('*//div[@class="pager-nav"]')[1].text.split()[3])

spending_dic={"Spending $ FY2022": [],
              "Type of Investment":[]
              }

link_dic={"Link": []
}

for page in range(1, pages + 1):
    print('Página {}. Linha {} de {}'.format(page, page, pages))
    print(page)
    while not re.search(f"Page {page} of", browser.find_elements_by_xpath('*//div[@class="pager-nav"]')[1].text):
        sleep(1)
        
    for elemento in browser.find_elements_by_xpath('*//span[contains(text(), "Spending $ FY2022")]//..'):
        if len(elemento.text.split('\n')) == 2:
            spending_dic["Spending $ FY2022"].append(elemento.text.split('\n')[1])
    for elemento in browser.find_elements_by_xpath('*//span[contains(text(), "Type of Investment")]//..'):
        if len(elemento.text.split('\n')) == 2:
            spending_dic["Type of Investment"].append(elemento.text.split('\n')[1])
            
    elems = browser.find_elements_by_xpath('*//span[@class="download-tag"]//a')
    links = [elem.get_attribute('href') for elem in elems]
    
    for link in links:
        link_dic["Link"].append(link)
        
            
    browser.find_element_by_xpath('*//a[@class="next"]').click()
            
df_links = pd.DataFrame(link_dic)
df_links['Check'] = '-'
df_links.to_csv('links.csv')
            
df_spending = pd.DataFrame(spending_dic)
writer = pd.ExcelWriter(download_path+'\\'+'Agencys and Spending.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
df_agencias.to_excel(writer, sheet_name='Agências')
df_spending.to_excel(writer, sheet_name='Individual Investments')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

```python
df_links = pd.read_csv('links.csv')
it = 1
for index, row in df_links.iterrows():
    print('baixando arquivo {}. Linha {} de {}'.format(index, it, len(df_links.index)))
    it = it+1
    if row["Check"] == 'Baixado':
        print('Arquivo já baixado')
        continue
    browser.get(row["Link"])
    browser.find_element_by_xpath('*//a[@class="print-link"]').click()
    df_links['Check'][index] = 'Baixado'
    df_links.to_csv('links.csv')
```
